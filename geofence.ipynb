```python
import pandas as pd
df = pd.read_csv('CPK_Parking_and_Restricted_Zones.csv')
df
```

```python
from point import Point
from polygon import Polygon
import pickle

geofence = df[df['PARKING_LOCATIONS'] == 'Geofence']['FENCECOORDINATES'][44]
geofence = geofence.split('\n')[3:-4]
count = 0

def chunks(lst, n):
    """Yield successive n-sized chunks from lst."""
    for i in range(0, len(lst), n):
        yield lst[i:i + n]

coords = []
for i in chunks(geofence, 4):
    latitude = i[1].lstrip().replace(',', '')
    longitude = i[2].lstrip()
    coords.append(Point(latitude, longitude))

coords = sorted(coords, key=lambda point: (point.lat, point.long))
```

```python
geofence = Polygon(coords)

with open('geofence.pickle', 'wb') as handle:
    pickle.dump(geofence, handle, protocol=pickle.HIGHEST_PROTOCOL)

with open('geofence.pickle', 'rb') as handle:
    unserialized_data = pickle.load(handle)

print(unserialized_data)
```
