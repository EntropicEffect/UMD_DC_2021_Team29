```python
from bs4 import BeautifulSoup
import requests
```

```python
class MapNode():
    def __init__(self, lattitude, longitude):
        self.lat = str(lattitude)
        self.long = str(longitude)
    
    def __repr__(self):
        return f'({self.lat}, {self.long})'
```

```python
## Retrieve Data from way webpage

way_url = "https://www.openstreetmap.org/api/0.6/way/488114799"
way_xml_data = requests.get(way_url).content
way_soup = BeautifulSoup(way_xml_data,"xml")
```

```python
## From way xml data scrape all node ids and store in node_id_list

node_set = way_soup.find_all('nd')
node_id_list = []
for node in node_set:
    node_id_list.append(node['ref'])
```

```python
# Extract latt/long from each node's xml data

node_list = []

for node_id in node_id_list:
    node_url = "https://www.openstreetmap.org/api/0.6/node/" + node_id
    node_xml_data = requests.get(node_url).content
    node_soup = BeautifulSoup(node_xml_data, "xml")
    node_data = node_soup.find("node")
    node_list.append(mapNode(node_data['lat'],node_data['lon']))
```
